### 1) ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
# Crear un array de 1000 elementos
bool_array = np.zeros(1000, dtype=bool)

# Obtener el tamaño en bytes
bytes_x_elemento = bool_array.itemsize

# multiplicando la cantidad de elementos por 
total_bytes = bool_array.size * bytes_x_elemento

print("Tamaño en bytes", total_bytes)
```

### 6) Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
# Obtener la lista de provincias
provincias = df['Provincia'].unique()

# Configurar el tamaños
num_provincias = len(provincias)
fig, axs = plt.subplots(num_provincias, figsize=(10, 5*num_provincias))

# Iterar y crear un gráficos con pandas
for i, provincia in enumerate(provincias):
    # Seleccionar los datos de la provincia actual
    datos_provincia = df[df['Provincia'] == provincia]
    municipios = datos_provincia['Municipio']
    poblaciones = datos_provincia['Poblacion']
    # Crear el gráfico en la posición adecuada
    axs[i].barh(municipios, poblaciones, color='blue')
    axs[i].set_title(provincia)
    axs[i].set_xlabel('Población')
    axs[i].set_ylabel('Municipio')

# Ajustar el diseño de la figura
plt.tight_layout()
plt.show()
```

### 8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python

# Coordenadas de Madrid
madrid_x, madrid_y = -3.7038, 40.4168

# Calcular la distancia entre Madrid y cada municipio
df['Distancia_a_Madrid'] = np.sqrt((df['X'] - madrid_x)**2 + (df['Y'] - madrid_y)**2)

# Excluir Madrid de los cálculos (distancia cero)
df_sin_madrid = df[df['Municipio'] != 'Madrid']

# Encontrar el municipio más cercano a Madrid
municipio_cercano = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmin()]

# Encontrar el municipio más lejano a Madrid
municipio_lejano = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmax()]


print("El municipio más cercano a Madrid es:", municipio_cercano['Municipio'])
print("Distancia:", municipio_cercano['Distancia_a_Madrid'], "kilómetros")

print("El municipio más lejano a Madrid es:", municipio_lejano['Municipio'])
print("Distancia:", municipio_lejano['Distancia_a_Madrid'], "kilómetros")
```
